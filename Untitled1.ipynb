```python
def normalize(x):
    x = x/255
    return x
```

```python
import glob
X_data = []
#For cats
files_cats = glob.glob("files/train/cats/*.jpg")
for myFile in files_cats:
    image = misc.imread(myFile)
    image = misc.imresize(image, (img_width, img_height))
    normalized_image = normalize(image)
    
    X_data.append(normalized_image)
```

```python
#For dogs
files_dogs = glob.glob("files/train/dogs/*.jpg")
for myFile in files_dogs:
    image = misc.imread(myFile)
    image = misc.imresize(image, (img_width, img_height))
    normalized_image = normalize(image)
    
    X_data.append(normalized_image)
```

```python
X_data = np.asarray(X_data)
X_data.shape
```

```python
Y_data = np.array([0] * 1024 + [1] * 1024)
Y_data
```

```python
img_width, img_height = 150, 150
batch_size = 32
```

```python
# a simple stack of 3 convolution layers with a ReLU activation and followed by max-pooling layers.
model = Sequential()
model.add(Convolution2D(32, (3, 3), input_shape=(img_width, img_height,3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Convolution2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Convolution2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.6))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(optimizer='rmsprop',
             loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
epochs = 20
```

```python
history = model.fit(X_data, Y_data, validation_split=0.30,  epochs = epochs)
#About 60 seconds an epoch when using CPU
```

```python
print(history.history.keys())
```

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
